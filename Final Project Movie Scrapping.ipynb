### Part 1: Scraping Movie Data

```python
from bs4 import BeautifulSoup
import requests
import os
import pandas as pd
import time
from collections import defaultdict, Counter
```

### Getting 50 more links of all movie genres

```python
genres = ["Adventure", "Animation", "Biography", "Comedy", "Crime", "Drama", "Family", "Fantasy", "Film-Noir", "History", 
          "Horror", "Music", "Musical", "Mystery", "Romance", "Sci-Fi", "Sport", "Thriller", "War", "Western"]

url_dict = {}

for genre in genres:
    url = "https://www.imdb.com/search/title/?genres={}&start=51&explore=title_type,genres&ref_=adv_nxt"
    formated_url = url.format(genre)
    url_dict[genre] = formated_url
    
print(url_dict)
```

```python
#dictionary of all the movies
movie_list_50 = []

#iterating through the genre links
for genre, url in url_dict.items():
    resp = requests.get(url)
    content = BeautifulSoup(resp.content, 'lxml')
    
    #scraping 50 movies per genre
    for movie in content.select('.lister-item-content'):
        time.sleep(1)
        try:
            movie_data = {
                        #movie title
                        'title':movie.select('.lister-item-header')[0].get_text().strip().replace('\n',''),
        
                        #movie year released
                        "year":movie.select('.lister-item-year')[0].get_text().strip(),
        
                        #parental advisory
                        'advisory':movie.select('.certificate')[0].get_text().strip(),
        
                        #movie length
                        'movie_length':movie.select('.runtime')[0].get_text().strip(),
        
                        #movie genre
                        'genre':movie.select('.genre')[0].get_text().strip(),
        
                        #movie rating
                        'rating':movie.select('.ratings-imdb-rating')[0].get_text().strip(),
        
                        #meta score of movie
                        'metascore':movie.select('.ratings-metascore')[0].get_text().strip(),
        
                        #movie description
                        'description':movie.select('.text-muted')[2].get_text().strip(),
        
                        #movie votes
                        'votes':movie.select('.sort-num_votes-visible')[0].get_text().strip()
                        }
            print('scraping movie ', movie.select('.lister-item-header')[0].get_text().strip().replace('\n',''))
        except IndexError:
            continue
            
        #adding movies to list
        movie_list_50.append(movie_data)
    
    #writing data to csv file
    dataframe = pd.DataFrame(movie_list_50)
    dataframe.to_csv('movies_dataset_50.csv')
print('Done.')
```
